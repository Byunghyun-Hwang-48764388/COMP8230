```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the motion dataset
df = pd.read_csv('motion_data.csv')

# Split the dataset into two categories for comparison: 
# 'medium' (representing daily movements) and 'heavy' (representing physical exercise).
# Select the first 100 samples.
daily_data = df[df['Category'] == 'medium'].head(100).reset_index(drop=True)
exercise_data = df[df['Category'] == 'heavy'].head(100).reset_index(drop=True)

# Part 1: Accelerometer Data Visualization ---
# Compare the acceleration patterns.
plt.figure(figsize=(12, 8))

# Subplot for Daily Activities
plt.subplot(2, 1, 1)
plt.plot(daily_data['Accelerometer_x'], label='Acc X', alpha=0.8)
plt.plot(daily_data['Accelerometer_y'], label='Acc Y', alpha=0.8)
plt.plot(daily_data['Accelerometer_z'], label='Acc Z', alpha=0.8)
plt.title('Accelerometer: Daily Life')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)

# Subplot for Exercise
plt.subplot(2, 1, 2)
plt.plot(exercise_data['Accelerometer_x'], label='Acc X', alpha=0.8)
plt.plot(exercise_data['Accelerometer_y'], label='Acc Y', alpha=0.8)
plt.plot(exercise_data['Accelerometer_z'], label='Acc Z', alpha=0.8)
plt.title('Accelerometer: Physical Exercise')
plt.xlabel('Index')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()

# Part 2: Gyroscope Data Visualization ---
# Compare angular velocity.
plt.figure(figsize=(12, 8))

# Subplot for Daily Activities
plt.subplot(2, 1, 1)
plt.plot(daily_data['Gyroscope_x'], label='Gyro X', alpha=0.8)
plt.plot(daily_data['Gyroscope_y'], label='Gyro Y', alpha=0.8)
plt.plot(daily_data['Gyroscope_z'], label='Gyro Z', alpha=0.8)
plt.title('Gyroscope: Daily Life')
plt.ylabel('Angular Velocity')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)

# Subplot for Physical Exercise
plt.subplot(2, 1, 2)
plt.plot(exercise_data['Gyroscope_x'], label='Gyro X', alpha=0.8)
plt.plot(exercise_data['Gyroscope_y'], label='Gyro Y', alpha=0.8)
plt.plot(exercise_data['Gyroscope_z'], label='Gyro Z', alpha=0.8)
plt.title('Gyroscope: Physical Exercise')
plt.xlabel('Index')
plt.ylabel('Angular Velocity')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
```
